# 2.1 Feature Selection

As we previously saw, the data consists of 67 features. Not all of them are important for our analysis, so we need to carefully select the ones that we may find useful.The first step is to **categorize** them as following:
- **Location and Date**
- **Covid cases**
- **Vaccinations Rate**
- **Demographics**
- **Medical and Life Quality**

```python
selected_cols = [
    
    # Location and Date
    "continent","location","date",
    
    # Covid cases
    "new_cases","new_cases_per_million","total_cases_per_million","new_deaths_per_million","total_deaths_per_million",
    "new_tests_per_thousand","reproduction_rate",
    
    # Vaccinations Rate
    "new_vaccinations_smoothed","people_vaccinated_per_hundred","total_boosters_per_hundred",
    
    # Demographics
    "population", "population_density","aged_65_older","aged_70_older",
    
    # Medical and Life Quality
    "handwashing_facilities","stringency_index","human_development_index","gdp_per_capita","extreme_poverty",
    "life_expectancy","hospital_beds_per_thousand"
    
]
```

```python
df_selected_data = df_data[selected_cols].copy()
df_selected_data.head(10)
```

# 3. Data Cleansing

One of the first and most important steps of **Data Analysis** is examining the acquired raw data and **_"clean"_** them. The outcome of this step will provide us with clean, quility data to work with.

```python
# View the data types of the selected features
df_selected_data.dtypes
```

## 3.1 Convert Date column to Datetime type

The first thing we see is that the **date** Feature which is stored as string. We will convert it to **Datetime** type

```python
# converts date to M8[ns], which is a 'synomym' to datetime64[ns]

df_selected_data['date']=pd.to_datetime(df_selected_data["date"],format='%Y-%m-%d')
df_selected_data["date"].dtypes
```

## 3.2 Dealing with NaN values

After finishing with the data types of the features we now have to deal with the **missing values** of our data. It is reasonable that not all countries have complete data, so we must deal with these issues. In the following table we see the count of our missing values.

```python
df_selected_data.isna().sum()
```

### 3.2.1 Removing unwanted NaN values

```python
'''
We observe that 11.126 rows have a missing value on the continent column.The reason is that there are data also for the entire
world, which we do not need in our analysis as it is country based.
We are going to solve this issue by removing the rows with missing values or whitespaces.
'''

# Replace whitespaces with NaN
df_selected_data.continent.replace('', pd.NA, inplace = True)

# Delete NaN values
df_selected_data.dropna(subset = ["continent"], inplace = True)
```

### 3.2.2 Filling NaN values with approximation data


There are also a small number of missing values in some of our depandand features like  "new_cases_per_million", "total_cases_per_million, "new_deaths_per_million", "total_deaths_per_million" . Performing interpolation in other columns with very high percentage of missing values, like "new_tests_per_thousand" or "people_vaccinated_per_hundred", may result in unreliable data that are not reflecting the reality. Also, these features are somewhat normal to have many missing values, as vaccines and tests came at a later stage of the covid outbreak.  <br><br>
This is something to be expected as not all countries provide daily data, especially the least developed ones.We need to fill those NaN values in a proper manner. One of the methods suitable to these kind of situations, where we have access to daily data, is by using the **interpolation method**. <br> <br>
Interpolation is the process of using known data values to estimate unknown data values, and on its simpliest forms is the **linear interpolation**. It is important to note that we make an approximate assumpion that the covid-19 follows a linear pattern. In the following code block we apply linear interpolation on specific columns. 


```python
# Sort our Dataframe by location in ascending order, and by date in Descending order, in order to property perform the
# interpolation

df_selected_data_sorted = df_selected_data.sort_values(["location", 'date'], ascending = [True, False]).copy()

df_selected_data_sorted.head(5)
```

```python
# the Feautures that we are going to perform interpolation

interpolate_list = ['new_cases_per_million', 'total_cases_per_million',
       'new_deaths_per_million', 'total_deaths_per_million','reproduction_rate']

# We perform interpolation on both directions(forward,backward) to fill all the missing values with the linear calculation

for column in interpolate_list:
    df_selected_data_sorted[column] = df_selected_data_sorted[column].interpolate(method = "linear",limit_direction = "both")
```
